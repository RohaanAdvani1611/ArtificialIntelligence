# 1. Solve the CSP of the N-queen problem:

#### The N Queen is the problem of placing N chess queens on an NÃ—N chessboard so that no two queens attack each other

```python
!pip install python-constraint
```

```python
from constraint import *
```

```python
p1 = Problem()
```

#### Enter N:

```python
N = int(input())
```

#### Generate Board:

```python
cols = range(N)
rows = range(N)
```

#### Add Variables:

```python
p1.addVariables(cols, rows)
```

#### Add the following Constraints:

1. No 2 Queens can be in the same Row.
2. No 2 Queens can be in the same Column.
3. No 2 Queens can be in the same Diagnol.

```python
for col1 in cols:
    for col2 in cols:
        if col1 < col2:
            p1.addConstraint(lambda row1, row2, col1=col1, col2=col2:
                                    abs(row1-row2) != abs(col1-col2) and
                                    row1 != row2, (col1, col2))
```

#### Generate Solution:

```python
solutions = p1.getSolutions()
```

# 2. Take any small CSP scheduling problem and generate all possible solutions:

#### Following problem is a CSP Time Table Scheduling Problem wherein user inputs the Number of Classes, Timings of Classes and Number of Professors. The Scheduling Solution is found such that at a particular time one Professor is in one Class only and all the Classes occur at scheduled time with some Professor available.

```python
p2 = Problem()
```

#### Enter Number of Classes:

```python
C = int(input())
```

```python
classes = range(C)
```

#### Enter start and end time of classes assuming college timings are 8 am - 12 pm (Enter only integer values):

```python
class_list = []
for i in range(C):
    print('Class no:',+i)
    st = int(input())
    et = int(input())
    list1 = [st, et]
    class_list.append(list1)
print(class_list)    
```

#### Enter Number of Professors:

```python
P = int(input())
```

```python
profs = range(P)
```

#### Add Variables:

```python
p2.addVariables(classes, profs)
```

#### Add Constraints:

```python
for class1 in classes:
    for class2 in classes:
        if class1 < class2:
            if((class_list[class1][0] > class_list[class2][0] and class_list[class1][0] < class_list[class2][1])
              or (class_list[class1][1] > class_list[class2][0] and class_list[class1][1] < class_list[class2][1])):
                p2.addConstraint(lambda prof1, prof2, class1=class1, class2=class2: prof1 != prof2, (class1, class2))
```

#### Generate Solution:

```python
solutions = p2.getSolutions()
```
